```python
# We import below all necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#the last line is just telling the notebook to include the figure in the page instead of opening a new window
```

```python
#Let's read the file and display some informations
#Note that "skiprows" force python to skip the first 2 lines of the
#file as they are not useful here
df = pd.read_csv('data_senegal.csv',skiprows=2)
df.info()
```

```python
#Here we split the dataset in two, creating two sub-dataset for 2000 and 2012
df2000 = df[df['Year']==2000]
df2012 = df[df['Year']==2012]
```

```python
#We can create a barchart to visualize the number of death per cause in 2000 by typing :
plt.bar(range(len(df2000['Both sexes'])),df2000['Both sexes'])
#To force the limits of the y-axis type :
plt.ylim(0.0,1600.0)
#To create a grid
plt.grid()
#To give a name to the y-axis
plt.ylabel('Deaths')
#To give a title to my figure
plt.title('Deaths per cause in 2000')
```

<h4>Some comments about the previous commands :</h4>
    * plt.ylim() and plt.ylabel() have an equivalent for the x-axis : plt.xlim() and plt.xlabel()
    * plt.grid() creates a grid following the ticks along both axes
    -> to change the ticks use the command : plt.xticks() and plt.yticks() giving between the brackets a list of value for where you want a tick.
    For instance try plt.xticks([]) or plt.xticks([0,25,50,75,100,125]) (by inserting them in the previous cell)

```python
#To do the same for the year 2012
plt.bar(range(len(df2012['Female'])),df2012['Female'])
plt.ylim(0.0,1600.0)
plt.grid()
plt.ylabel('Deaths')
plt.title('Deaths per cause in 2012')
```

```python
plt.figure(figsize=(12,7))
#The previous command just force the size of my new figure

#If we want to visualize everything in teh same figure but separated plots we use "plt.subplot"
#Let's work on the first subplot :
plt.subplot(211)
plt.bar(range(len(df2000['Both sexes'])),df2000['Both sexes'])
plt.ylim(0.0,1600.0)
plt.grid()
plt.ylabel('Deaths')
plt.title('Deaths per cause in 2000')

#We can now work on the second subplot (notice the difference in subplot):
plt.subplot(212)
plt.bar(range(len(df2012['Both sexes'])),df2012['Both sexes'])
plt.ylim(0.0,1600.0)
plt.grid()
plt.ylabel('Deaths')
plt.title('Deaths per cause in 2012')

# You can adjust the spacing between subplots by using : plt.subplots_adjust()
```

<h4>In the following we will create a figure of the number of deaths in 2012 vs the number of deaths in 2000.</h4>

```python
#First create a new figure with a specific size (necessary if you are using a terminal)
plt.figure(figsize=(9,9))

#Before creating any figure we define what X and Y are
X = df2000['Both sexes'].values
Y = df2012['Both sexes'].values

#plot the figure using the "scatter" function
#plt.scatter(X,Y)
#Note that you can change the marker type via the "marker" keyword. For instance try :
plt.scatter(X,Y,marker='^')

#We now give the same scale and limits to the x-axis and the y-axis
plt.axis('scaled')
plt.xlim(0.0,1000.0)
plt.ylim(0.0,1000.0)

#Then let's create the y=x line
plt.plot([0.0,1000.0],[0.0,1000.0],color='k')

#we can also add a grid
plt.grid()

#but more important we need to label our axes ...
plt.xlabel('Deaths in 2000')
plt.ylabel('Deaths in 2012')
#... and give a title (not necessary if you use a caption)
plt.title('Deaths in 2012 vs in 2000')
plt.savefig('deaths2012vs2000.png')
```
